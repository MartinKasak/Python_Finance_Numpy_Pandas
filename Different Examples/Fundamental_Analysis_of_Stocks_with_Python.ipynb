```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
yf.Ticker("AAPL").info
```

```python
tickers = ["AAPL", "MSFT", "DPZ"]
```

```python
infos = []

for i in tickers:
    infos.append(yf.Ticker(i).info)
```

```python
infos
```

```python
fundamentals = ["dividendYield", "marketCap", "beta", "forwardPE"]
```

```python
df = pd.DataFrame(infos)
```

```python
df = df.set_index('shortName')
```

```python
df
```

```python
df[df.columns[df.columns.isin(fundamentals)]]
```

```python
plt.bar(df.index, df.dividendYield, color=("k", "orange", "cyan"))
plt.show()
```

```python
plt.bar(df.index, df.forwardPE, color=("k", "orange", "cyan"))
plt.show()
```
