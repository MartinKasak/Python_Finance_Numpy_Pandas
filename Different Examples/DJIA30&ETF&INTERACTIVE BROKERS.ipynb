```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
url = "https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average"
url
```

```python
df = pd.read_html(url)[1]
df
```

```python
df.rename(columns = {"Date added":"Date_Added"}, inplace = True)
df.rename(columns = {"Index weighting":"Weights"}, inplace = True)
```

```python
df.Date_added = pd.to_datetime(df.Date_Added)
```

```python
df.Weights = pd.to_numeric(df.Weights.str.replace("%", ""))
```

```python
df
```

```python
df.Weights.sum()
```

```python
df.drop(columns = "Notes", inplace = True)
```

```python
df.set_index("Symbol", inplace = True)
```

```python
df
```

```python
symbols = df.index.to_list()
symbols
```

```python
last_update = df["Date_Added"].max()
last_update
```

```python
index = "^GSPC"
index_tr = "^SP500TR"
etf = "SPY"
```

```python
df = yf.download([index, index_tr, etf], end = "2022-11-30")
df
```

```python
df.dropna(inplace = True)
```

```python
prices = df.Close.copy()
prices
```

```python
norm = prices / prices.iloc[0]
norm
```

```python
norm[["SPY", "^GSPC"]].plot(figsize = (12, 8))
plt.title("S&P500: ETF Price vs. Price Return Index", fontsize = 15)
plt.show()
```

```python
norm.loc["2022",["SPY", "^SP500TR"]].plot(figsize = (12, 8))
plt.title("S&P500: ETF TR vs. Price Return Index", fontsize = 15)
plt.show()
```

```python
returns = prices.pct_change()
returns
```

```python
def ann_risk_return(returns_df):
  summary = pd.DataFrame(index = returns_df.columns)
  summary["ann. Risk"] = returns_df.std() * np.sqrt(252)
  log_returns = np.log(returns_df + 1)
  summary["CAGR"] = np.exp(log_returns.mean()*252)-1
  return summary
```
