Normality of financial returns

---



```python
import numpy as np
dpz["log_ret"] = np.log(dpz.Close / dpz.Close.shift())
dpz
```

```python
dpz.describe()
```

```python
dpz.log_ret.plot(kind= "hist", figsize = (15, 8), bins = 100, fontsize = 15, density = False)
plt.xlabel("Daily Returns", fontsize = 15)
plt.ylabel("Frequency", fontsize = 15)
plt.title("Frequancy Distribution of Returns", fontsize = 20)
plt.show()
```

```python
mu = dpz.log_ret.mean()
mu
```

```python
sigma = dpz.log_ret.std()
sigma
```

```python
import scipy.stats as stats
```

```python
stats.skew(dpz.log_ret.dropna())
```

```python
stats.kurtosis(dpz.log_ret.dropna(), fisher = True)
```

```python
x = np.linspace(dpz.log_ret.min(), dpz.log_ret.max(), 10000)
x
```

```python
y = stats.norm.pdf(x, loc = mu, scale = sigma)
y
```

```python
plt.figure(figsize = (20, 8))
plt.hist(dpz.log_ret, bins = 500, density = True, label ="Frequancy Distribution of Returns" )
plt.plot(x, y , linewidth = 3, color = "red", label = "Normal distribution")
plt.title("Normal distribution", fontsize = 15)
plt.xlabel("Daily Returns", fontsize = 15)
plt.ylabel("pdf", fontsize = 15)
plt.legend(fontsize = 15)
plt.show()
```

```python
z_stat, p_value = stats.normaltest(dpz.log_ret.dropna())
```

```python
z_stat
```

```python
p_value
```

```python
returns = close.apply(lambda x: np.log(x.dropna() / x.dropna().shift()))
returns
```

```python
returns.info()
```

```python
summary = returns.agg(["mean", "std"]).T
summary
```

```python
summary.columns = ["Mean", "Std"]
summary
```

```python
summary.plot(kind = "scatter", x = "Std", y = "Mean", figsize = (15, 12), s = 50, fontsize = 15)
for i in summary.index:
  plt.annotate(i, xy=(summary.loc[i, "Std"]+ 0.00005, summary.loc[i, "Mean"]+0.00005), size = 15)
plt.xlabel("Rist(std)", fontsize = 15)
plt.ylabel("Mean Return", fontsize = 15)
plt.title("Mean-Variance Analysis", fontsize = 20)
plt.show()
```
